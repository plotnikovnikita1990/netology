##Задание 1
Для датафрейма log из материалов занятия создайте столбец source_type по правилам:

- если источник traffic_source равен Yandex или Google, то в source_type ставится organic;
- для источников paid и email из России ставим ad;
- для источников paid и email не из России ставим other;
- все остальные варианты берём из traffic_source без изменений.

```python
import pandas as pd
```

```python
log = pd.read_csv('visit_log.csv', sep=';')
log.head()
```

```python
log['source_type'] = log['traffic_source'] # берём из traffic_source без изменений
```

```python
log.loc[log['traffic_source'].isin(['yandex', 'google']), 'source_type'] = 'organic' # если источник traffic_source равен Yandex или Google, то в source_type ставится organic
```

```python
log.loc[log['traffic_source'].isin(['paid', 'email']) & (log['region'] == 'Russia'), 'source_type'] = 'ad' # для источников paid и email из России ставим ad;
```

```python
log.loc[log['traffic_source'].isin(['paid', 'email']) & (log['region'] != 'Russia'), 'source_type'] = 'other' # для источников paid и email не из России ставим other
```

##Задание 3
Используйте файл с оценками фильмов ml-latest-small/ratings.csv. Посчитайте среднее время жизни пользователей, которые выставили более 100 оценок.
Под временем жизни понимается разница между максимальным и минимальным значениями столбца timestamp для данного значения userId.

```python
data_ratings = pd.read_csv('ratings.csv')
data_ratings.head()
```

```python
# сгруппирую по userId	добавлю столбцы с count по rating и min/max timestamp
ratings_stats = data_ratings.groupby('userId').agg(rating_count=('rating', 'count'), min_timestamp=('timestamp', 'min'), max_timestamp=('timestamp', 'max')) # в agg можно сразу добавлять столбцы с вычеслениями
ratings_stats.head()
```

```python
# отфильтрую пользователей с count > 100
users_100 = ratings_stats.loc[ratings_stats['rating_count'] > 100] # без .loc появляется предупреждение
users_100.head()
```

```python
# найду разницу между максимальным и минимальным значениями столбца timestamp
users_100.loc[:, 'lifetime'] = users_100['max_timestamp'] - users_100['min_timestamp']
users_100.head()
```

```python
# в более читаемом формате
users_100.loc[:, 'date'] = pd.to_timedelta(users_100['lifetime'], unit='s')
users_100.head()
```

Дана статистика услуг перевозок клиентов компании по типам (см. файл “Python_13_join.ipynb” в разделе «Материалы для лекции “Продвинутый pandas”» ---- Ноутбуки к лекции «Продвинутый pandas»).
Нужно сформировать две таблицы:

- таблицу с тремя типами выручки для каждого client_id без указания адреса клиента;
- аналогичную таблицу по типам выручки с указанием адреса клиента.

Обратите внимание, что в процессе объединения таблиц данные не должны теряться.

### К домашнему заданию №4
Дана статистика услуг перевозок клиентов компании по типам:
- rzd - железнодорожные перевозки
- auto - автомобильные перевозки
- air - воздушные перевозки
- client_base - адреса клиентов

```python
rzd = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115],
        'rzd_revenue': [1093, 2810, 10283, 5774, 981]
    }
)
rzd
```

```python
auto = pd.DataFrame(
    {
        'client_id': [113, 114, 115, 116, 117],
        'auto_revenue': [57483, 83, 912, 4834, 98]
    }
)
auto
```

```python
air = pd.DataFrame(
    {
        'client_id': [115, 116, 117, 118],
        'air_revenue': [81, 4, 13, 173]
    }
)
air
```

```python
client_base = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115, 116, 117, 118],
        'address': ['Комсомольская 4', 'Энтузиастов 8а', 'Левобережная 1а', 'Мира 14', 'ЗЖБИиДК 1',
                    'Строителей 18', 'Панфиловская 33', 'Мастеркова 4']
    }
)
client_base
```

```python
# таблица с тремя типами выручки для каждого client_id без указания адреса клиента
joined_revenue = client_base[['client_id']]
joined_revenue = joined_revenue.merge(rzd, on='client_id', how='left')
joined_revenue = joined_revenue.merge(auto, on='client_id', how='left')
joined_revenue = joined_revenue.merge(air, on='client_id', how='left')
joined_revenue
```

```python
# таблица по типам выручки с указанием адреса клиента
joined_address = joined_revenue.merge(client_base, on='client_id', how='left')
joined_address
```
