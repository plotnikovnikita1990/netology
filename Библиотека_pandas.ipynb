Задание 1
Определите, какому фильму было выставлено больше всего оценок 5.0.

```python
import pandas as pd
```

```python
data_movies = pd.read_csv('movies.csv')
data_movies.head()
```

```python
data_ratings = pd.read_csv('ratings.csv')
data_ratings.head()
```

```python
# отфильтрую таблицу data_ratings по рейтингу 5.0
data_ratings_5 = data_ratings[ (data_ratings['rating']==5.0)]
data_ratings_5.head()
```

```python
# топ movieId по количеству оценок 5.0
top_rating = data_ratings_5.groupby('movieId').count().sort_values(by="rating", ascending=False)
top_rating.head()
```

```python
# как вариант вместо группировки можно использовать более короткую версию с методом value_counts:
data_ratings_5.movieId.value_counts()
```

```python
# объединю таблицу топ рейтинга с фильмами и вывожу топ 1
joined = top_rating.merge(data_movies, on='movieId', how='left')
joined.head(1)
```

Задание 2
По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 год. Не учитывайте в расчётах отрицательные значения quantity.

```python
data_power = pd.read_csv('power.csv')
data_power.head()
```

```python
# оказывается литва  - 'Lithuania'
data_power[ data_power['country'].str.contains('li', case=False) ]['country'].unique()
```

```python
# можно немного сократить код, используя для фильтра стран метод isin:
data_power['country'].isin(['Latvia', 'Estonia', 'Lithuania'])
```

```python
# фильтрую таблицу по странам
filtered_countries = data_power[(data_power['country']=='Latvia') | (data_power['country']=='Lithuania') | (data_power['country']=='Estonia')]

filtered_countries.head()
```

```python
# фильтрую по условию категория 4, 12 и 21
filtered_countries_category = filtered_countries[(filtered_countries['category'] == 4) | (filtered_countries['category'] == 12) | (filtered_countries['category'] == 21)]
filtered_countries_category.head()
```

```python
# фильтрую по условию период с 2005 по 2010 год и исключаю отрицательные значения quantity
filtered_countries_year = filtered_countries_category[(filtered_countries_category['year'].between(2005, 2010)) & (filtered_countries_category['quantity'] > 0)]
filtered_countries_year.head()
```

```python
# суммирую по quantity
sum_power = filtered_countries_year['quantity'].agg('sum')
sum_power

# ответ в формате np.float64(240580.0)
```
